Load RAW Data from MySQL

```python
import pandas as pd
import mysql.connector
import seaborn as sns
import matplotlib.pyplot as plt

conn = mysql.connector.connect(
    host = "localhost",
    user ="root",
    password = "newpassword123",
    db = "black_friday_db"
)
df  = pd.read_sql("SELECT * FROM black_friday_raw", conn)
df.head
```

Handling Missing Values

```python
# Product Categories
df["Product_Category_2"].fillna(0,inplace = True)
df["Product_Category_3"].fillna(0,inplace = True)

# missing = product doesnot belong to category
# acts as a valid place holder
```

Encode Categorical variables

```python
df["Gender"]= df["Gender"].map({"M":1, "F":0})
```

Age (Ordinal Encoding)

```python
age_map = {
    "0-17":0,
    "18-25":1,
    "26-35":2,
    "36-45":3,
    "46-50":4,
    "51-55":5,
    "55+":6
}
df["Age"] = df["Age"].map(age_map)
```

City Category

```python
df["City_Catergory"] = df["City_Category"].map({"A":0,"B":1,"C":2})
```

Stay in Current City

```python
df["Stay_In_Current_City_Years"] = df["Stay_In_Current_City_Years"].replace("4+",4).astype(int)
```

Drop Unnecessary Columns

```python
df.drop(["User_ID","Product_ID"],axis=1, inplace = True)
```
